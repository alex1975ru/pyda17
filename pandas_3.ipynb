```python
import pandas as pd
df = pd.read_csv('C:\\Users\\kuznecov.ap\\Desktop\\piton\\13 Продвинутый Pandas\\visit_log.csv', sep = ';')
df.head(20)
def source_type(row):
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        return "organic"
    elif row['traffic_source'] == 'paid' and row['region'] == 'Russia' or row['traffic_source'] == 'email' and row['region'] == 'Russia' :
        return "ad"
    elif row['traffic_source'] == 'paid' or row['traffic_source'] == 'email':
        return "other"
    else:
        return row['traffic_source']
df['source_type'] = df.apply(source_type, axis=1)
df.head()
```

```python
df = pd.read_csv('C:\\Users\\kuznecov.ap\\Desktop\\piton\\13 Продвинутый Pandas\\ml-latest-small\\ratings.csv', sep = ',')
df_sorted = df.groupby('userId').count().sort_values('userId', ascending=True).query('rating>100')
data = pd.read_csv('C:\\Users\\kuznecov.ap\\Desktop\\piton\\13 Продвинутый Pandas\\ml-latest-small\\ratings.csv', names = ['userId', 'movieId1', 'rating1', 'timestamp1'], sep = ',', header=0)
data_j = data.merge(df_sorted, how='left', on='userId')
data_j_s = data_j.query('rating>100')
ratings_grouped = data_j_s.groupby('userId').agg([min, max])
ratings_grouped['diff'] = ratings_grouped['timestamp1']['max'] - ratings_grouped['timestamp1']['min']
ratings_grouped['diff'].mean()
ratings_grouped.head()[['diff']]/24/3600
```
