```python
q1 = new_df.surgical_lesion.quantile(0.25)
q3 = new_df.surgical_lesion.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = new_df[new_df.surgical_lesion.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
# print(lower_bound, upper_bound)
```

# Задание 3. Работа с пропусками


```python
new_df.info()
```

```python
for col in new_df.columns:
    pct_missing = new_df[col].isna().mean()
    print(f'{col} - {pct_missing :.1%}')
```

Предполагаем, что пропуски носят случайный характер, поэтому предлагаю для числовых переменных произвести замену на медиану а для категориальных на моду (хотя в лекции Вы не рекомендовали так делать, но на данном этапе трудно выявить взаимосвязь переменных между собой).
- Age (Возраст): Пропусков нет 
- rectal_temperature (Ректальная температура): числовая переменная и 20,1% пропусков- заменяем на медиану
- pulse (Пульс): числовая переменная и 8% пропусков- заменяем на медиану
- respiratory_rate (Частота дыхания): числовая переменная и 19,4% пропусков- заменяем на медиану
- pain (боль):  категориальная переменная и 18,4% пропусков- заменяем на моду
- packed_cell_volume (объем упакованных клеток): числовая переменная и 9,7% пропусков- заменяем на медиану
- outcome (гибель): категориальная переменная и 0,3% пропусков- заменяем на моду
- surgical_lesion? (хирургическое повреждение?): Пропусков нет

```python
new_df.rectal_temperature.fillna(new_df.rectal_temperature.median(), inplace=True)
new_df.rectal_temperature.fillna(new_df.rectal_temperature.median()).isna().sum()
new_df.pulse.fillna(new_df.pulse.median(), inplace=True)
new_df.pulse.fillna(new_df.pulse.median()).isna().sum()
new_df.respiratory_rate.fillna(new_df.respiratory_rate.median(), inplace=True)
new_df.respiratory_rate.fillna(new_df.respiratory_rate.median()).isna().sum()
new_df.pain.fillna(new_df.pain.mode()[0], inplace=True)
new_df.packed_cell_volume.fillna(new_df.packed_cell_volume.median(), inplace=True)
new_df.packed_cell_volume.fillna(new_df.packed_cell_volume.median()).isna().sum()
new_df.outcome.fillna(new_df.outcome.mode()[0], inplace=True)
```

```python
new_df.rectal_temperature.fillna(new_df.rectal_temperature.median()).isna().sum()
```
