# Уровень 0
Посчитайте значение первой и второй производной функции:
𝑓(𝑥)=𝑥5+4𝑠𝑖𝑛(2𝑥)+𝑐𝑜𝑠(3𝑥+3)

В точке x = 1.

```python
import matplotlib.pyplot as plt
import numpy as np
import pylab
from scipy.misc import derivative
```

```python
def f(x):
    return x**5 + 4*(np.sin(2*x)) + np.cos(3*x+3)
derivative(f, 1.0, dx=1e-6) #первая производная
```

```python
derivative(f, 1.0, dx=1e-6, n=2) #вторая производная
```

# Уровень 1:
Постройте график первой и второй производной функции:

𝑓(𝑥)=𝑠𝑖𝑛5(2𝑥+1) На отрезке [-5; 5].

```python
x = np.linspace(-5, 5)
fx = f(x)
def f(x):
    return np.sin(2*x+1)**5

fx = [derivative(f, xx, dx=1e-6, n=1) for xx in x] #первая производная
```

```python
f1x = [derivative(f, xx, dx=1e-6, n=2) for xx in x] #вторая производная
```

```python
plt.plot(x, fx, label='первая производная')
plt.legend()
plt.show()
```

```python
plt.plot(x, f1x, label='вторая производная')
plt.legend()
plt.show()
```

```python
plt.plot(x, fx, label='первая производная')
plt.plot(x, f1x, label='вторая производная')
plt.legend()
plt.show()
```
