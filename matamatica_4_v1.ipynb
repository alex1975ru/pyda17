# 햒쮏쒫왏쫨 0
쮐혢햦혝햟햧혝햣 향햫햟혢햣햫햦햣 햣쒫쮏 햦 쒬쮐쮏 쮏쟳쒫쮏얧쫧쮏 혟혞햫햨혡햦햦:
洧녭(洧논)=洧논5+4洧멇롐뒳롐(2洧논)+洧녫洧녶洧(3洧논+3)

 혝쮐햨햣 x = 1.

```python
import matplotlib.pyplot as plt
import numpy as np
import pylab
from scipy.misc import derivative
```

```python
def f(x):
    return x**5 + 4*(np.sin(2*x)) + np.cos(3*x+3)
derivative(f, 1.0, dx=1e-6) #햣쒫썜 쮏쟳쒫쮏얧쫧썜
```

```python
derivative(f, 1.0, dx=1e-6, n=2) #쒬쮐햟혪 쮏쟳쒫쮏얧쫧썜
```

# 햒쮏쒫왏쫨 1:
쮐혝쮏좜햣 햡햟혟햦햨 햣쒫쮏 햦 쒬쮐쮏 쮏쟳쒫쮏얧쫧쮏 혟혞햫햨혡햦햦:

洧녭(洧논)=洧멇롐뒳롐5(2洧논+1) 햏햟 쮐햣향햨햣 [-5; 5].

```python
x = np.linspace(-5, 5)
fx = f(x)
def f(x):
    return np.sin(2*x+1)**5

fx = [derivative(f, xx, dx=1e-6, n=1) for xx in x] #햣쒫썜 쮏쟳쒫쮏얧쫧썜
```

```python
f1x = [derivative(f, xx, dx=1e-6, n=2) for xx in x] #쒬쮐햟혪 쮏쟳쒫쮏얧쫧썜
```

```python
plt.plot(x, fx, label='햣쒫썜 쮏쟳쒫쮏얧쫧썜')
plt.legend()
plt.show()
```

```python
plt.plot(x, f1x, label='쒬쮐햟혪 쮏쟳쒫쮏얧쫧썜')
plt.legend()
plt.show()
```

```python
plt.plot(x, fx, label='햣쒫썜 쮏쟳쒫쮏얧쫧썜')
plt.plot(x, f1x, label='쒬쮐햟혪 쮏쟳쒫쮏얧쫧썜')
plt.legend()
plt.show()
```
