```python
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
# создаем три вектора
x = np.array([1, 1])
y = np.array([2, 0])
z = np.array([0, 2])
```

```python
import matplotlib.pyplot as plt
```

# 1. Изобразить вектора на экране

```python
plt.figure()
plt.plot(x, "o")
plt.plot(y, "o")
plt.plot(z, "o")
plt.grid();
```

```python
ax = plt.axes()

plt.xlim( [0, 2.5] )
plt.ylim( [0, 2.5] )

ax.arrow( 0, 0, x[0], x[1], head_width=0.1, head_length=0.2, fc='k', ec='k' )
ax.arrow( 0, 0, y[0], y[1], head_width=0.1, head_length=0.2, fc='k', ec='k' )
ax.arrow( 0, 0, z[0], z[1], head_width=0.1, head_length=0.2, fc='k', ec='k' )
plt.grid()
plt.show()
```

# 3. Найти угол между векторами x и y

```python
from numpy import linalg
# длины векторов
xLength = np.linalg.norm( x )
yLength = np.linalg.norm( y )
cosine = np.dot( x, y ) / ( xLength * yLength )
```
